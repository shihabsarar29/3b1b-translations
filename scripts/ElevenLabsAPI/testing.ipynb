```python
import dotenv
import os

# Load the environment variables from the .env file
dotenv.load_dotenv()

# Get the API key from the environment
api_key = os.environ.get("ELEVEN_LABS_API_KEY")
```

```python
import requests

CHUNK_SIZE = 1024

url = "https://api.elevenlabs.io/v1/text-to-speech/{voice_id}".format(voice_id="tx65YVj8FpeLCD8e467Y")

payload = {
    "text": "Hey, I am Grant Sanderson, and this is the Shark Tank."
}

headers = {"Content-Type": "application/json", "xi-api-key": api_key, "Accept": "audio/mpeg"}

response = requests.request("POST", url, json=payload, headers=headers)

print(response.text)

with open('output.mp3', 'wb') as f:
    for chunk in response.iter_content(chunk_size=CHUNK_SIZE):
        if chunk:
            f.write(chunk)

" Success "
```
